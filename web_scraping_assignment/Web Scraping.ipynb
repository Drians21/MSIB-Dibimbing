```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd 
```

```python
url = 'https://www.petsecure.com.au/pet-care/a-guide-to-worldwide-pet-ownership/'
response = requests.get(url)

print(response)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')

#print(soup.prettify())
```

```python
cats = soup.find('div', {'class': 'table-holder clearfix'})
```

```python
table = cats.find('table', {'class':'cats'})

print(table.prettify())
```

```python
x = table.find_all('th')
for i, y in enumerate(x):
  print(i, y)
  print(i, "text:", y.text)
```

```python
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())

print(headers)
```

```python
x = table.find_all('tr')
for i, y in enumerate(x):
  print(i, y.prettify())
  print(i, "text:", y.text)
```

```python
data = []
for tr in table.find_all('tr'):
    row_data = []
    for td in tr.find_all('td'):
        row_data.append(td.text.strip())
    if len(row_data) > 0:
        data.append(row_data)
```

```python
data
```

```python
df = pd.DataFrame(data, columns = headers)
df
```

## Cleansing/Transform

```python
df.info()
```

```python
df = df.rename(columns={
        'TOP 20\nCAT POPULATIONS':'country',
        '':'total_population'
})
```

```python
df.info()
```

```python
df['total_population'] = df['total_population'].str.replace(',', '')
df['total_population'] = df['total_population'].astype(int)
```
